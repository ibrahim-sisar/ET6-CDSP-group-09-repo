```python
import pandas as pd
```

```python
raw_data = pd.read_csv("../1_datasets/raw_datasets/covid_deaths.csv")
cleaned_death_data = raw_data.copy()
cleaned_death_data.head()
```

```python
cleaned_death_data.columns
```

```python
cleaned_death_data.shape
```

```python
cleaned_death_data.isnull().sum()
```

```python
cleaned_death_data = cleaned_death_data.drop(["sex", "age", 'measure', 'cause'], axis=1, errors="ignore")
cleaned_death_data = cleaned_death_data.rename(
    columns={"val": "COVID Deaths Val", "location": "Country"}
)
cleaned_death_data.head()
```

```python
countries_to_keep = [
    "Egypt",
    "India",
    "United States of America",
    "Chile",
    "Nigeria",
    "Italy",
    "China",
    "Japan",
    "Afghanistan",
    "Germany",
    "Russian Federation",
    "Spain",
    "Romania",
    "Indonesia",
    "Saudi Arabia",
    "Bangladesh",
    "Republic of Korea",
    "South Africa",
    "Ethiopia",
    "Kenya",
    "Brazil",
    "Mexico",
    "Canada",
    "Australia",
    "Fiji",
]
cleaned_death_data = cleaned_death_data[
    cleaned_death_data["Country"].isin(countries_to_keep)
]
cleaned_death_data.head(20)
```

```python
cleaned_death_data.reset_index(drop=True)
```

```python
cleaned_death_data.to_csv("cleaned_death.csv", index=False)
```
